## 5.4 Predicciones para el conjunto de competición

Una vez entrenado y almacenado el modelo final, lo utilizamos para generar predicciones sobre el conjunto de datos de la competición. Dado que este conjunto no dispone de etiquetas, en esta fase no se realiza una evaluación del rendimiento, sino que simplemente se aplica el modelo aprendido para obtener la clase predicha en cada observación.

Por ello, en este segundo cuaderno, nos limitamos a cargar el archivo simulado de competición, aplicar el modelo previamente guardado (que ya contiene toda la lógica de transformación de preprocesado gracias a la pipeline) y volcar los resultados en un CSV.

```python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import FunctionTransformer

def transform_pdays(X):
    X_out = X.copy()
    X_out['wascontacted'] = X_out['pdays'].apply(lambda x: 0 if x == -1 else 1)
    
    def bin_pdays(x):
        if x == -1: return "no_contacto"
        elif x < 100: return "0_99"
        elif x < 200: return "100_199"
        elif x < 400: return "200_399"
        else: return "400_plus"
    
    X_out['pdays_binned'] = X_out['pdays'].apply(bin_pdays)
    X_out = X_out.drop(columns=['pdays'])
    return X_out

# Ahora sí podemos cargar el modelo
modelo = joblib.load('modelo_final.joblib')
print("Modelo cargado correctamente.")

# Cargamos los datos de competición
df_competicion = pd.read_pickle("bank_competition.pkl")
print(f"Shape del dataset de competición: {df_competicion.shape}")

# Corregimos None en job y marital
df_competicion['job']     = df_competicion['job'].replace({None: np.nan})
df_competicion['marital'] = df_competicion['marital'].replace({None: np.nan})

# Generamos predicciones
predicciones = modelo.predict(df_competicion)

print(f"\nDistribución de predicciones:")
print(pd.Series(predicciones).value_counts())

# Guardamos en CSV
df_predicciones = pd.DataFrame(predicciones, columns=['prediccion'])
df_predicciones.to_csv('predicciones.csv', index=False)

print(f"\nPredicciones guardadas en 'predicciones.csv' ({len(predicciones)} filas).")
print(df_predicciones.head(10))
```

### Conclusión: Predicciones de la competición

La generación de predicciones sobre el conjunto de datos de la competición se ha realizado correctamente a partir del modelo final entrenado. Gracias al uso de una pipeline, el conjunto de competición recibe el mismo tratamiento que los datos empleados durante el entrenamiento, incluyendo tanto la transformación de la variable pdays como el resto de etapas de preprocesado. Esto permite aplicar el modelo de forma consistente y reduce el riesgo de errores derivados de diferencias en el tratamiento de los datos.

Previamente a la inferencia, recordamos que fue necesario homogeneizar los valores nulos de algunas variables, sustituyendo los valores None por np.nan en las columnas correspondientes, para mantener el mismo formato utilizado durante el desarrollo del modelo.

Finalmente, las predicciones obtenidas se han almacenado correctamente en el fichero predicciones.csv, cumpliendo así con lo solicitado en el enunciado para la entrega del conjunto de competición.

# 6. Comparación de predicciones: App vs Modelo
En este apartado abordamos el despliegue del modelo en producción mediante una aplicación web desarrollada con Streamlit. El objetivo es disponer de una herramienta sencilla que permita a los gestores del banco introducir los valores de un nuevo cliente y obtener automáticamente la predicción generada por el modelo entrenado. 

Para ello, la aplicación carga el fichero modelo_final.joblib, que contiene la pipeline completa seleccionada en los apartados anteriores. De este modo, cualquier nueva instancia introducida en la aplicación recibe exactamente el mismo tratamiento que los datos utilizados durante el entrenamiento, incluyendo las transformaciones, el preprocesado y la clasificación final.

Con el fin de verificar que el despliegue funciona correctamente, hemos considerado dos nuevas instancias y hemos comparado las predicciones obtenidas directamente mediante la pipeline cargada en Python con las generadas por la aplicación web en Streamlit.

```python
import joblib
import pandas as pd

# Cargar modelo final
model = joblib.load("modelo_final.joblib")

# Dos nuevas instancias
instancias = pd.DataFrame([
    {
        'age': 35,
        'job': 'admin.',
        'marital': 'married',
        'education': 'secondary',
        'default': 'no',
        'balance': 1000,
        'housing': 'yes',
        'loan': 'no',
        'contact': 'cellular',
        'day': 15,
        'month': 'may',
        'duration': 300,
        'campaign': 2,
        'pdays': -1,
        'previous': 0,
        'poutcome': 'unknown'
    },
    {
        'age': 52,
        'job': 'management',
        'marital': 'single',
        'education': 'tertiary',
        'default': 'no',
        'balance': 5200,
        'housing': 'no',
        'loan': 'no',
        'contact': 'telephone',
        'day': 8,
        'month': 'nov',
        'duration': 520,
        'campaign': 1,
        'pdays': 180,
        'previous': 2,
        'poutcome': 'success'
    }
], index=["Instancia 1", "Instancia 2"])

# Predicciones
instancias["prediccion"] = model.predict(instancias)

# Mostrar tabla final
print(instancias)
```

En ambos casos, las predicciones obtenidas coinciden exactamente.Esto nos confirma que el despliegue se ha realizado correctamente y que la aplicación web utiliza de forma adecuada la misma pipeline entrenada, sin alterar el flujo de transformación ni el proceso de predicción. Por tanto, la app permite realizar predicciones consistentes para nuevos clientes, cumpliendo con el objetivo planteado.
